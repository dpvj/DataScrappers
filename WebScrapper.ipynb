```python
L=[]

for j in range(1,100):
    driver = webdriver.Chrome(options=options, executable_path='chromedriver')
    stealth(driver, languages=["en-US", "en"], vendor="Google Inc.", platform="Win32",webgl_vendor="Intel Inc.", renderer="Intel Iris OpenGL Engine", fix_hairline=True)
    driver.get('https://www.iproperty.com.my/sale/kuala-lumpur/all-residential/?page='+str(j))
    counter=0
    for i in range(1,30):
        try:
            location = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div/ul/li['+str(i)+']/div/div/div[3]/div[2]/div/h2').text
            area = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div/ul/li['+str(i)+']/div/div/div[3]/div[2]/div/div[1]').text
            details = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div/ul/li['+str(i)+']/div/div/div[3]/div[2]/div/div[2]/p').text
            price = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div/ul/li['+str(i)+']/div/div/div[3]/div[1]/ul/li').text
            date = driver.find_element(By.XPATH, '/html/body/div/div/div[3]/div/ul/li['+str(i)+']/div/div/div[1]/div[1]/p').text
            agent = driver.find_element(By.XPATH, '/html/body/div/div/div[3]/div/ul/li['+str(i)+']/div/div/div[1]/div[1]/span/div[2]').text
            agency = driver.find_element(By.XPATH, '/html/body/div/div/div[3]/div/ul/li['+str(i)+']/div/div/div[1]/div[2]/div/img').get_property('attributes')[0].get('value')
            link = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div/ul/li['+str(i)+']/div/div/div[4]/div/button[2]/a').get_property('href')
            rooms = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div/ul/li['+str(i)+']/div/div/div[3]/div[3]/div[1]/ul/li[1]').text
            bathroom = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div/ul/li['+str(i)+']/div/div/div[3]/div[3]/div[1]/ul/li[2]').text
            try:
                parking = driver.find_element(By.XPATH, '//*[@id="app"]/div/div[3]/div/ul/li['+str(i)+']/div/div/div[3]/div[3]/div[1]/ul/li[3]').text
            except:
                parking = np.nan
            L.append((location, area, details, price, date, agent, agency, link, rooms, bathroom, parking))
            counter += 1
        except:
            pass
    time.sleep(10)
    print(f'Page {j} done, scrapped {i} records, obtained {counter} records, total of {len(L)} records')
    driver.close()
    time.sleep(10)

driver.quit()
```

```python
df = pd.DataFrame(L, columns=['location', 'area', 'details', 'price', 'date', 'agent', 'agency', 'link', 'rooms', 'bathroom', 'parking'])
df
```
